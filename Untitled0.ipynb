```python
import json

class Student: 12
    def __init__(self, mssv, name, age, major):
        self.mssv = mssv
        self.name = name
        self.age = age
        self.major = major

    def to_dict(self):
        return {
            "MSSV": self.mssv,
            "Họ tên": self.name,
            "Tuổi": self.age,
            "Ngành": self.major
        }

class StudentManager:
  # Nếu chưa có dữ liệu thì tạo dữ liệu mẫu
if not manager.students:
    sample_students = [
        Student("SV001", "Nguyễn Văn A", 20, "Khoa học dữ liệu"),
        Student("SV002", "Trần Thị B", 21, "Công nghệ thông tin"),
        Student("SV003", "Lê Văn C", 19, "Kỹ thuật phần mềm"),
    ]
    for s in sample_students:
        manager.add_student(s)
    print("Đã thêm dữ liệu mẫu.")

    def __init__(self, filename="students.json"):
        self.filename = filename
        self.students = self.load_students()

    def load_students(self):
        try:
            with open(self.filename, "r", encoding="utf-8") as f:
                return [Student(**data) for data in json.load(f)]
        except (FileNotFoundError, json.JSONDecodeError):
            return []

    def save_students(self):
        with open(self.filename, "w", encoding="utf-8") as f:
            json.dump([s.to_dict() for s in self.students], f, ensure_ascii=False, indent=4)

    def add_student(self, student):
        self.students.append(student)
        self.save_students()

    def show_students(self):
        if not self.students:
            print("Chưa có sinh viên nào!")
        else:
            for s in self.students:
                print(f"{s.mssv} - {s.name}, {s.age} tuổi, Ngành: {s.major}")

    def find_student(self, mssv):
        for s in self.students:
            if s.mssv == mssv:
                return s
        return None

    def remove_student(self, mssv):
        student = self.find_student(mssv)
        if student:
            self.students.remove(student)
            self.save_students()
            print("Đã xóa sinh viên:", mssv)
        else:
            print("Không tìm thấy sinh viên!")

def menu():
    manager = StudentManager()

    while True:
        print("\n===== QUẢN LÝ SINH VIÊN =====")
        print("1. Thêm sinh viên")
        print("2. Hiển thị danh sách")
        print("3. Tìm sinh viên theo MSSV")
        print("4. Xóa sinh viên")
        print("5. Thoát")
        choice = input("Chọn chức năng: ")

        if choice == "1":
            mssv = input("Nhập MSSV: ")
            name = input("Nhập họ tên: ")
            age = int(input("Nhập tuổi: "))
            major = input("Nhập ngành: ")
            manager.add_student(Student(mssv, name, age, major))
            print("Thêm sinh viên thành công!")
        elif choice == "2":
            manager.show_students()
        elif choice == "3":
            mssv = input("Nhập MSSV cần tìm: ")
            student = manager.find_student(mssv)
            if student:
                print("Thông tin:", student.to_dict())
            else:
                print("Không tìm thấy sinh viên!")
        elif choice == "4":
            mssv = input("Nhập MSSV cần xóa: ")
            manager.remove_student(mssv)
        elif choice == "5":
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ!")
```
